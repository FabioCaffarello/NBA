```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_{}_totals.html'

#Method to extract the information of players per season
def scrape_season_stats(base_url, year_start, year_end):
    years = range(year_start,year_end+1,1)

    df_season = pd.DataFrame()

    for year in years:
        print(f'Extracting data from the {year} season')
        request_url = base_url.format(year)
        req = requests.get(request_url)
        soup = BeautifulSoup(req.content, 'html.parser')
        table = soup.find('table', {'id':'totals_stats'})
        df = pd.read_html(str(table))[0]
        df['Year'] = year
        df['Season'] = f'{year-1}-{str(year)[2:]}'
        df_season = df_season.append(df)
    return df_season
```

```python
#Using theMethod to extract the information of players per season
df_season = scrape_season_stats(url, 1985, 2020)
```

```python
#Cleaning from the data extracted in the method to use them
df_season_clean = df_season[~(df_season['Rk'] == 'Rk')]
df_season_clean
```

```python
#Writing CSV of season stats
df_season_clean.to_csv('season_stats.csv')
```
