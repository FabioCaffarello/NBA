# IMPORTS

## Libraries

```python
import pandas as pd
```

## Load Data

```python
dfMVP = pd.read_csv('../data/all_MVP.csv')
dfSeasons = pd.read_csv('../data/season_stats.csv')
```

# Data Preparation

## Renane Columns - DfMVP

```python
dfMVP.rename(columns={'Player':'MVP'}, inplace=True)
```

## Clean Player Column of DfSeasons

```python
dfSeasons['Player'] = dfSeasons['Player'].apply(lambda row: row.strip('*'))
```

## Merge DfSeasons and DfMVP

```python
df = pd.merge(left = dfSeasons, right = dfMVP[['Season','MVP']], how='left', left_on=['Season', 'Player'], right_on=['Season', 'MVP'])
```

## Convert to .csv

```python
df.to_csv('../data/all_Seasons_MVP.csv', index=False)
```
